```python
def create_plot(mongo, elastic,title):
    plt.figure()
    plt.plot(mongo['Latency'].tolist(),label = "MongoDB Performance")
    plt.plot(elastic['Latency'].tolist(), label = "Elasticsearch Performance")
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
create_plot(mongo_insert, elastic_insert,"Database Insert Performance")
```

```python
create_plot(mongo_find, elastic_find, "Database Find Performance")
```

```python
create_plot(mongo_wildcard, elastic_wildcard,"Database Wildcard Search Performance")
```

```python
create_plot(mongo_update, elastic_update,"Database Update Performance")
```
